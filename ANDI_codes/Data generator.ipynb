```python
# Functions needed from andi_datasets
import numpy as np
import pandas as pd
from andi_datasets.datasets_challenge import challenge_phenom_dataset, _get_dic_andi2, _defaults_andi2
```

```python
# Setting parameter of trajec

MODELS = np.arange(5)
# MODELS = np.arange(1)
NUM_FOVS = 30
# PATH = 'transformer (data)/data (6.17 (k == 3))/'   # Chose your path!

dics = []


for m in MODELS:   
    dic = _get_dic_andi2(m+1)

            # Fix length and number of trajectories 
    dic['T'] = 200
    dic['N'] = 400

            #### SINGLE STATE ####
        
#         k range
# v1: ((1.0,0.5))
# v2: ((2.0,1.0))
# v3: ((3.0,1.5))
# v4: ((6.0,2.5))

    if m == 0:  
        dic['alphas'] = np.array([1.0, 1.0])
        dic['Ds'] = np.array([1.0, 0.5])
    
            
    ### MULTI STATE ####
    if m == 1:
#         3-state model with different alphas 
        
#     k range
# v1: ((0.5,0.3), (1.0,0.3), (1.5,0.3))
# v2: ((1.0,0.5), (1.5,0.5), (2.5,0.5))
# v3: ((1.5,1.0), (2.5,1.0), (3.5,1.5))
# v4: ((2.0,1.0), (3.5,1.5), (5.0,2.5))

        dic['Ds'] = np.array([[0.5, 0.3],
                              [1.0, 0.3],
                              [1.5, 0.3]])

        dic['alphas'] = np.array([[0.5, 0.3],
                                  [1.0, 0.3],
                                  [1.5, 0.3]])

        dic['M'] = np.array([[0.98, 0.01, 0.01],
                             [0.01, 0.98, 0.01],
                             [0.01, 0.01, 0.98]]) 
        

        
    #### IMMOBILE TRAPS ####
    if m == 2:
        dic['Ds'] = np.array([1.0, 0.5])
        dic['alphas'] = np.array([1.0, 0.5])
        dic['r'] = np.array([2])
        dic['Nt'] = 100
        dic['Pb'] = 0.8
        dic['Pu'] = 0.2
            
    
    #### DIMERIZATION ####
    if m == 3:
        dic['Ds'] = np.array([[1.5, 0.5],
                              [0.3, 0.2]])

        dic['alphas'] = np.array([[1.5, 0.3],
                                  [0.5, 0.3]])
        dic['Pb'] = 0.98
        dic['Pu'] = 0.02
        
        
    ## CONFINEMENT ####
    if m == 4:
        dic['Ds'] = np.array([[1.0, 1.0],
                              [1.0, 1.0]])

        dic['alphas'] = np.array([[1.0, 0.0]]),
        
        dic['trans'] = 0.2
        
        dic['Nc'] = 50
        
        dic['r'] = 5

#     # Add repeated fovs for the experiment
    for _ in range(NUM_FOVS):
        dics.append(dic)
```

```python


import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt

# data = pd.read_csv('../' + 'Andi challenge2/' + PATH + 'trajs_exp_0_fov_0.csv')


dirnames = [PATH]

csv_model = {}

n = 2000
for i,dirname in enumerate(dirnames):
    filenames = os.listdir(dirname)

    
#     csv_zip = ['trajs_exp_0_fov_0.csv']
    csv_single = []
    csv_multi = []
    csv_immobile = []
    csv_dimerization = []
    csv_confinement = []
    
    for filename in filenames:
        if os.path.splitext(filename)[1] == '.csv':
            txt = filename

            txt = txt.split("_")
            if int(txt[2]) <= 30:
                if len(csv_single) < n:
                    csv_single.append(filename)
            
            if 30 <= int(txt[2]) <= 60:
                if len(csv_multi) < n:
                    csv_multi.append(filename)
                    
            if 60 <= int(txt[2]) <= 90:
                if len(csv_immobile) < n:
                    csv_immobile.append(filename)
                    
            if 90 <= int(txt[2]) <= 120:
                if len(csv_dimerization) < n:
                    csv_dimerization.append(filename)
                    
            if 120 <= int(txt[2]) <= 150:
                if len(csv_confinement) < n:
                    csv_confinement.append(filename)
                    
csv_model['single'] = csv_single
csv_model['multi'] = csv_multi
csv_model['immobile'] = csv_immobile
csv_model['dimerization'] = csv_dimerization
csv_model['confinement'] = csv_confinement

                

models = ['single', 'multi', 'immobile', 'dimerization', 'confinement']


data = {}
data_single = []
data_multi = []
data_immobile = []
data_dimerization = []
data_confinement = []

for model_num,model in enumerate(csv_model):
    data_model = []
    if model_num == 0:
        for num,file in enumerate(csv_model[model]):
            
#                
            data_single.append(pd.read_csv(dirnames[0] + file))
            
    if model_num == 1:
        for num,file in enumerate(csv_model[model]):
            
            data_multi.append(pd.read_csv(dirnames[0] + file))
            
    if model_num == 2:
        for num,file in enumerate(csv_model[model]):
           
            data_immobile.append(pd.read_csv(dirnames[0] + file))
            
    if model_num == 3:
        for num,file in enumerate(csv_model[model]):
            
            data_dimerization.append(pd.read_csv(dirnames[0] + file))
            
    if model_num == 4:
        for num,file in enumerate(csv_model[model]):
            
            data_confinement.append(pd.read_csv(dirnames[0] + file))
            
        
```

```python
models = ['single', 'multi', 'immobile', 'dimerization', 'confinement']
# models = ['confinement']
from andi_datasets.utils_challenge import label_continuous_to_list

# data = {}
for model_num,list_model in enumerate(list_data):
    data_model = []
    
    for num,pd_data in enumerate(list_model):
        
        idxs = np.unique(pd_data['traj_idx'].iloc[3:].values.astype('int') )
        data_TID = pd_data.groupby(['traj_idx'])
        data_n = [[],[],[],[],[],[]]
        for idx in idxs:
            tmp = data_TID[['frame','x','y','alpha','D','state']].get_group(idx).\
                                astype('float').sort_values(by='frame').to_numpy().transpose()

            if models[model_num] == 'single':
#                 if len(tmp[0]) == T and len(np.unique(tmp[3])) == 1:
                if len(np.unique(tmp[3])) == 1 and len(np.unique(tmp[4])) == 1:                
                    data_n[0].append(tmp[1] - tmp[1][0])
                    data_n[1].append(tmp[2] - tmp[2][0])
                    data_n[2].append(tmp[0])
                    data_n[3].append(tmp[3])
                    data_n[4].append(tmp[4])
                    data_n[5].append(tmp[5])
                    
            if models[model_num] == 'multi':
#                 if len(tmp[0]) == 200 and len(np.unique(tmp[3])) >= 2 and 0 not in np.asarray(np.unique(tmp[5])) and 1 not in np.asarray(np.unique(tmp[5])):
                if len(np.unique(tmp[3])) >= 2 or len(np.unique(tmp[4])) >= 2:
                    data_n[0].append(tmp[1] - tmp[1][0])
                    data_n[1].append(tmp[2] - tmp[2][0])
                    data_n[2].append(tmp[0])
                    data_n[3].append(tmp[3])
                    data_n[4].append(tmp[4])
                    data_n[5].append(tmp[5])
                    
            if models[model_num] == 'immobile':
                if len(np.unique(tmp[3])) >= 2 or len(np.unique(tmp[4])) >= 2:
                    data_n[0].append(tmp[1] - tmp[1][0])
                    data_n[1].append(tmp[2] - tmp[2][0])
                    data_n[2].append(tmp[0])
                    data_n[3].append(tmp[3])
                    data_n[4].append(tmp[4])
                    data_n[5].append(tmp[5])
                    
            if models[model_num] == 'dimerization':
                if len(np.unique(tmp[3])) >= 2 or len(np.unique(tmp[4])) >= 2:
                    data_n[0].append(tmp[1] - tmp[1][0])
                    data_n[1].append(tmp[2] - tmp[2][0])
                    data_n[2].append(tmp[0])
                    data_n[3].append(tmp[3])
                    data_n[4].append(tmp[4])
                    data_n[5].append(tmp[5])
                    
            if models[model_num] == 'confinement':
                if len(np.unique(tmp[4])) >= 2:
                    data_n[0].append(tmp[1] - tmp[1][0])
                    data_n[1].append(tmp[2] - tmp[2][0])
                    data_n[2].append(tmp[0])
                    data_n[3].append(tmp[3])
                    data_n[4].append(tmp[4])
                    data_n[5].append(tmp[5])
        if len(data_n[0]) > 0:
            data_model.append(data_n)
    data[models[model_num]] = data_model
```
